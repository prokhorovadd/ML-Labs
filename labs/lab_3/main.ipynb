```python
import pandas as pd
from heapq import nlargest
import statsmodels.formula.api as sm
from sklearn.metrics import classification_report
from sklearn.utils.multiclass import type_of_target
import seaborn as sb
```

```python
df = pd.read_csv('Boston.csv')

df.head()
```

```python
corr = df.corr()

corr['medv'] = corr['medv'].map(abs)
res = nlargest(2, corr['medv'])

display(corr)
display(res)
```

```python
model = sm.ols('medv ~ lstat', df).fit()

print(model.summary())
```

```python
sb.regplot(data=df, x="medv", y="lstat")
```

```python
model = sm.ols('medv ~ dis + crim + zn + indus + chas + nox + rm + age + dis + rad + tax + ptratio + black + lstat', df, subset=None, drop_cols=None).fit()
print(model.summary())
```
