```python
import numpy as np
from sklearn.datasets import load_wine
import pandas as pd
from heapq import nlargest
import statsmodels.formula.api as sm
from sklearn.metrics import mean_squared_error, median_absolute_error, classification_report
```

```python
data = load_wine()
df = pd.DataFrame(data=np.c_[data['data'], data['target']], columns=data['feature_names'] + ['target']) 

display(df)
```

```python
df = df.loc[df['target'] != 2]

display(df)
```

```python
correlation = df.corr()
correlation['target'] = correlation['target'].map(abs)
res = nlargest(4, correlation['target'])

display(res)
display(correlation)
```

Больше всего с target коррелируют proline, color_intensity, alcohol

```python
model = sm.logit('target ~ proline + color_intensity + alcohol', df, subset=None, drop_cols=None).fit()
print(model.summary())
```

```python
true = df.loc[:, 'target']
predict = model.predict(df.loc[:, ['proline', 'color_intensity', 'alcohol']]) 

display(predict)
```

```python
true = df.loc[:, 'target']
predict = model.predict(df.loc[:, ['proline', 'color_intensity', 'alcohol']]) 

errors = {'MSE': [mean_squared_error(true, predict)], 'MAE': [median_absolute_error(true, predict)]}
display(pd.DataFrame(data=errors))
print(classification_report(true, predict))
```
